```python
import importlib
import functions as func
import functions_vectorized as func_vec
import numpy as np
import time
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

importlib.reload(func)
importlib.reload(func_vec)

np.set_printoptions(precision=10, suppress=True)

N = [2 ** i for i in range(1, 14)]

time1 = np.array([])
time2 = np.array([])
time3 = np.array([])

for n in N:
    X = np.random.random((n,2))
    Y = np.random.random((n,2))
    start = time.time()
    func.pairwise_distance(X.tolist(), Y.tolist())
    end = time.time()

    start_vec = time.time()
    func_vec.pairwise_distance(X, Y)
    end_vec = time.time()

    start_sc = time.time()
    cdist(X, Y)
    end_sc = time.time()

    time1 = np.append(time1, end - start)
    time2 = np.append(time2, end_vec - start_vec)
    time3 = np.append(time3, end_sc - start_sc)

plt.figure(figsize=(6, 6))

plt.plot(N, time1, marker="o", linestyle='-', color="black", label="Невекторизованная")
plt.plot(N, time2, marker="o", linestyle='-', color="red", label="Векторизованная")
plt.plot(N, time3, marker="o", linestyle='-', color="yellow", label="Scipy")

plt.title("График")
plt.xlabel("Размер массивов X и Y")
plt.ylabel("Время (c)")
plt.grid()
plt.legend()
plt.tight_layout()

plt.show()
```
